# 1. Data structure

## **Definition:** In Python, data structures are fundamental ways to organize and store collections of data. Think of them as specialized containers designed to handle data in a specific manner, making it efficient to access, manage, and manipulate. Choosing the appropriate data structure is crucial for writing effective and performant code, especially when dealing with the often large and complex datasets encountered in machine learning.

### Different data structures are like different organizational systems. A dictionary is like a phone book where you can quickly look up information by name. A list is like a to-do list where items are in a specific order. A set is like a club membership list where each person can only be included once.

# 2. List

image.png

## **Definition:** A list is an ordered, mutable collection of items that can contain elements of different data types, including other lists. Lists are defined by enclosing items in square brackets [] with elements separated by commas.

Key Characteristics:
1. Ordered: Items have a defined position (index).
2. Mutable: Items can be added, removed, or modified.
3. Heterogeneous: Can hold mixed data types (e.g., [1, "apple", True]).
4. Dynamic: Size adjusts automatically as items change.


# Creating Lists

```
# Empty list
empty_list = []

# List with elements of the same type
numbers = [1, 2, 3, 4, 5]

# List with mixed data types
mixed_list = [1, "Hello", 3.14, True]

# Nested lists
nested = [[1, 2], [3, 4], [5, 6]]

# Creating a list using the list() function
characters = list("Python")  # Result: ['P', 'y', 't', 'h', 'o', 'n']

```



# Accessing List Elements


```
fruits = ["apple", "banana", "cherry", "date", "elderberry"]

# Indexing (starts at 0)
first_fruit = fruits[0]  # Result: "apple"
third_fruit = fruits[2]  # Result: "cherry"

# Negative indexing (counts from the end)
last_fruit = fruits[-1]  # Result: "elderberry"
second_last = fruits[-2]  # Result: "date"

# Slicing [start:stop:step]
first_three = fruits[0:3]  # Result: ["apple", "banana", "cherry"]
every_other = fruits[::2]  # Result: ["apple", "cherry", "elderberry"]
reversed_list = fruits[::-1]  # Result: ["elderberry", "date", "cherry", "banana", "apple"]
```

# Modifying Lists
```
colors = ["red", "green", "blue"]

# Changing an element
colors[1] = "yellow"  # Result: ["red", "yellow", "blue"]

# Adding elements
colors.append("purple")  # Result: ["red", "yellow", "blue", "purple"]
colors.insert(2, "orange")  # Result: ["red", "yellow", "orange", "blue", "purple"]
colors.extend(["pink", "black"])  # Result: ["red", "yellow", "orange", "blue", "purple", "pink", "black"]

# Removing elements
colors.remove("yellow")  # Removes the first occurrence of "yellow"
popped_color = colors.pop()  # Removes and returns the last element ("black")
popped_index = colors.pop(1)  # Removes and returns element at index 1 ("orange")
del colors[0]  # Removes element at index 0 ("red")
colors.clear()  # Empties the list completely
```

# List Operations

```
# Concatenation
list1 = [1, 2, 3]
list2 = [4, 5, 6]
combined = list1 + list2  # Result: [1, 2, 3, 4, 5, 6]

# Repetition
repeated = list1 * 3  # Result: [1, 2, 3, 1, 2, 3, 1, 2, 3]

# Length
length = len(combined)  # Result: 6

# Membership testing
is_present = 5 in combined  # Result: True
is_absent = 7 in combined  # Result: False

# Min and max
minimum = min(combined)  # Result: 1
maximum = max(combined)  # Result: 6

# Sum
total = sum(combined)  # Result: 21

```

# List Methods and Functions

image.png

```
numbers = [3, 1, 4, 1, 5, 9, 2, 6]

# Sorting
numbers.sort()  # Modifies the list in-place: [1, 1, 2, 3, 4, 5, 6, 9]
sorted_copy = sorted(numbers)  # Returns a new sorted list without modifying the original

# Reversing
numbers.reverse()  # Modifies the list in-place: [9, 6, 5, 4, 3, 2, 1, 1]

# Counting occurrences
count_of_ones = numbers.count(1)  # Result: 2

# Finding index
index_of_five = numbers.index(5)  # Result: 2 (index of the first occurrence)

# Copying
shallow_copy = numbers.copy()  # Creates a shallow copy
another_copy = numbers[:]  # Another way to create a shallow copy
```

# List Comprehensions
```
# Creating lists with a compact syntax
squares = [x**2 for x in range(1, 6)]  # Result: [1, 4, 9, 16, 25]

# With conditional filtering
even_squares = [x**2 for x in range(1, 11) if x % 2 == 0]  # Result: [4, 16, 36, 64, 100]

# Nested list comprehension
matrix = [[i+j for j in range(3)] for i in range(3)]
# Result: [[0, 1, 2], [1, 2, 3], [2, 3, 4]]
```

# Common Patterns with Lists

```
# Finding the sum of all elements
numbers = [1, 2, 3, 4, 5]
total = sum(numbers)  # Result: 15

# Finding average
average = sum(numbers) / len(numbers)  # Result: 3.0

# Filtering
evens = [num for num in numbers if num % 2 == 0]  # Result: [2, 4]

# Transforming
doubled = [num * 2 for num in numbers]  # Result: [2, 4, 6, 8, 10]

# Finding min/max index
min_index = numbers.index(min(numbers))  # Result: 0
max_index = numbers.index(max(numbers))  # Result: 4
```

# 3. Dictionary

image.png

## **Definition:** A dictionary is an unordered, mutable collection of key-value pairs. Each key must be unique and immutable (like strings, numbers, or tuples), while values can be of any data type and duplicated. Dictionaries are defined using curly braces {} with key-value pairs separated by colons.


# Creating Dictionaries

```
# Empty dictionary
empty_dict = {}
another_empty = dict()

# Dictionary with key-value pairs
student = {"name": "John", "age": 20, "grade": "A"}

# Using dict() constructor
person = dict(name="Alice", age=25, country="USA")

# Dictionary with different data types
mixed = {
    "text": "hello",
    1: [1, 2, 3],
    (1, 2): "tuple key",
    True: "boolean key"
}

# Nested dictionaries
contacts = {
    "John": {"phone": "555-1234", "email": "john@example.com"},
    "Mary": {"phone": "555-5678", "email": "mary@example.com"}
}
```

# Accessing Dictionary Elements
```
student = {"name": "John", "age": 20, "grade": "A"}

# Accessing by key
name = student["name"]  # Result: "John"

# Using get() method (safer, returns None if key doesn't exist)
age = student.get("age")  # Result: 20
address = student.get("address")  # Result: None
address = student.get("address", "Not provided")  # Result: "Not provided" (default value)

# Getting all keys, values, and items
all_keys = student.keys()  # Result: dict_keys(['name', 'age', 'grade'])
all_values = student.values()  # Result: dict_values(['John', 20, 'A'])
all_items = student.items()  # Result: dict_items([('name', 'John'), ('age', 20), ('grade', 'A')])

# Checking if a key exists
has_name = "name" in student  # Result: True
has_address = "address" in student  # Result: False
```

# Modifying Dictionaries

```
user = {"username": "jsmith", "active": True}

# Adding or updating values
user["email"] = "john@example.com"  # Adds new key-value pair
user["username"] = "johnsmith"  # Updates existing value

# Updating with another dictionary
user.update({"phone": "555-1234", "active": False})  # Adds new and updates existing

# Removing items
removed_email = user.pop("email")  # Removes key and returns value
removed_item = user.popitem()  # Removes and returns the last inserted item as tuple
del user["phone"]  # Removes the key-value pair

# Clearing all items
user.clear()  # Empties the dictionary
```

# Dictionary Methods

```
inventory = {"apples": 5, "bananas": 10, "oranges": 8}

# Creating a copy
inventory_copy = inventory.copy()

# Getting value with default if key doesn't exist
grapes = inventory.get("grapes", 0)  # Result: 0

# Setting default value if key doesn't exist
inventory.setdefault("pears", 0)  # Adds "pears": 0 if "pears" doesn't exist

# Creating a new dictionary with keys from sequence and same value
new_stock = dict.fromkeys(["strawberries", "cherries"], 0)
# Result: {"strawberries": 0, "cherries": 0}
```

# Dictionary Comprehensions

```
# Creating a dictionary from another collection
names = ["Alice", "Bob", "Charlie"]
name_lengths = {name: len(name) for name in names}
# Result: {"Alice": 5, "Bob": 3, "Charlie": 7}

# Filtering with conditions
numbers = range(1, 6)
squares = {n: n**2 for n in numbers if n % 2 == 0}
# Result: {2: 4, 4: 16}

# Transforming keys and values
stock = {"apples": 5, "bananas": 10, "oranges": 8}
uppercase_stock = {k.upper(): v*2 for k, v in stock.items()}
# Result: {"APPLES": 10, "BANANAS": 20, "ORANGES": 16}
```

# Common Dictionary Patterns
```
# Counting occurrences of items
words = ["apple", "banana", "apple", "orange", "banana", "apple"]
word_count = {}
for word in words:
    word_count[word] = word_count.get(word, 0) + 1
# Result: {"apple": 3, "banana": 2, "orange": 1}

# Better way using Counter from collections
from collections import Counter
word_count = Counter(words)
# Result: Counter({"apple": 3, "banana": 2, "orange": 1})

# Grouping data
students = [
    {"name": "Alice", "grade": "A"},
    {"name": "Bob", "grade": "B"},
    {"name": "Charlie", "grade": "A"}
]
by_grade = {}
for student in students:
    grade = student["grade"]
    if grade not in by_grade:
        by_grade[grade] = []
    by_grade[grade].append(student["name"])
# Result: {"A": ["Alice", "Charlie"], "B": ["Bob"]}
```

# Nested Dictionary Operations
```
contacts = {
    "John": {"phone": "555-1234", "email": "john@example.com"},
    "Mary": {"phone": "555-5678", "email": "mary@example.com"}
}

# Accessing nested values
john_email = contacts["John"]["email"]  # Result: "john@example.com"

# Adding nested information
contacts["John"]["address"] = "123 Main St"

# Adding a new person with complete information
contacts["Bob"] = {"phone": "555-4321", "email": "bob@example.com"}
```
