# Histograms






We can perform univariate analysis on these columns based on their datatype:

    For numerical data, we can use a histogram to visualize the data distribution.
    
    The number of bins should be chosen appropriately to represent the data well.

    For categorical data, we can use a bar plot to visualize the frequency of each category.



````python
# Create histogram
plt.figure(figsize=(15, 5))
plt.hist(data['month'], bins=12, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Month')
plt.xlabel('Month of Flight')
plt.ylabel('Frequency')
plt.grid(alpha=0.3)
plt.show()
````

### **When to use:**
- For continuous numerical data to understand distribution shapes
- To identify the range, central tendency, and potential outliers
- When you need to understand data dispersion and skewness

# Bar Charts (for categorical data)






````python
plt.figure(figsize=(15,5))
counts = data['year'].value_counts()
counts.plot(kind='bar', edgecolor='white')
plt.title('Year of Flight')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.grid(alpha=0.3, axis='y')
plt.show()
````

````python
# A three-letter code signifying the airport from which the flight departed.

plt.figure(figsize=(15,5))
counts = data['origin'].value_counts()
counts.plot(kind='bar', edgecolor='white')
plt.title('Origin')
plt.xlabel('Departure Airport Code')
plt.ylabel('Frequency')
plt.grid(alpha=0.3, axis='y')
plt.show()
````

### **When to use:**
- For categorical data to show counts or frequencies
- When comparing values across different categories
- For displaying results of groupby operations

# 2. Bivariate Analysis

## **Definition:** Bivariate analysis examines the relationship between two variables to detect associations, correlations, or cause-and-effect relationships.

# Violin Plots

````python

# Create violin plot
plt.figure(figsize=(15, 5))
sns.violinplot(x = data['month'], y = data['arr_delay'])
# sns.violinplot(x=x, y=y, palette=colors_box)
plt.title('Month vs Arrival Delay')
plt.xlabel('Month')
plt.ylabel('Arrival Delay')
plt.grid(alpha=0.3)
plt.show()
````

### **When to use:**
- For comparing distributions across groups
- When the distribution shape is important

# Scatter Plot

````python
plt.figure(figsize=(10, 6))
plt.scatter(x= data['dep_delay'], y = data['arr_delay'], alpha=0.5)
plt.title('Departure Delay vs Arrival Delay')
plt.xlabel('Departure Delay')
plt.ylabel('Arrival Delay')
plt.grid(alpha=0.3)
plt.show()
````

### **When to use:**
- For examining relationships between two numerical variables
- To identify correlations, clusters, or outliers
- When looking for patterns that might suggest causality

# 3. Multivariate Analysis

## **Definition:** Multivariate analysis examines relationships among three or more variables simultaneously to uncover complex patterns, clusters, and interactions.

````python
# For correlation heatmap, let's consider the numeric features only
numeric_features = ['month', 'dep_time', 'dep_delay', 'hour']
correlation_matrix = data[numeric_features].corr()

plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True,  annot_kws={"size": 8})
plt.xticks(fontsize=7)
plt.yticks(fontsize=7)
plt.title('Correlation Heatmap', fontsize=15)
plt.show()
````
