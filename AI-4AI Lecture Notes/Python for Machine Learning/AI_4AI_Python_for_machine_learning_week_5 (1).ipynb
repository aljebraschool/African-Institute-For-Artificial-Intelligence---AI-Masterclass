## **Definition:** Data preprocessing is the process of transforming raw data into a clean, structured format suitable for analysis and machine learning. It involves handling missing values, standardizing features, removing outliers, encoding categorical variables, and other techniques that prepare data for modeling.

## Why Data Preprocessing?
  - Improves model accuracy and reliability
  - Reduces training time and computational requirements
  - Helps identify patterns and relationships in data
  - Prevents algorithm bias from unbalanced or skewed data
  - Ensures consistent formatting and scales across features

# 3. Handling Missing Values

## **Why Handle Missing Values?**
- Most machine learning algorithms cannot handle missing values natively, and they can introduce bias or errors in the analysis.




# **Techniques for Handling Missing Values:**

# Deletion Approaches: remove rows with missing values

### **When to use:** When missing data is minimal and randomly distributed, or when specific columns have too many missing values.

```
data = data.dropna()
```



````python
data = data.dropna()
````

````python
# check missing value again to confirm

print(data.isnull().sum())
````

# Imputation Approaches:


### **When to use:**
- Mean: For normally distributed data without outliers
- Median: For skewed numerical data or data with outliers
- Median: For skewed numerical data or data with outliers

```
# Impute with mean, median, or mode
df_mean = df.copy()
df_mean['numeric_col'].fillna(df['numeric_col'].mean(), inplace=True)

df_median = df.copy()
df_median['numeric_col'].fillna(df['numeric_col'].median(), inplace=True)

df_mode = df.copy()
df_mode['categorical_col'].fillna(df['categorical_col'].mode()[0], inplace=True)
```

image.png

# Advanced Imputation:

### **When to use:** When simple imputation methods would disrupt relationships between features, or when more accurate imputation is needed.



```
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_value =np.nan,  strategy = 'mean')
#provided your data is in numpy values
data['columns'] = imputer.fit_transform(data['columns']) # apply it on the columns you're interested in
```

# 6. Feature Encoding

## **Why Encode Features?**
- Machine learning algorithms work with numerical data, so categorical variables need conversion.




# **Encoding Techniques:**

# Label Encoding (for ordinal categories)

```
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
label_encoder = LabelEncoder()
data['categorical_column'] = label_encoder.fit_transform(data['categorical_column'])
```



# One-Hot Encoding (for nominal categories)

````python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
#create the object of the encoder class
onehot_encoder = OneHotEncoder(sparse_output=False)

#reshape the intended column into two dimension
origin_column_reshapred = data['origin'].values.reshape(-1, 1)

#now encode the origin column
origin_column_encoded = onehot_encoder.fit_transform(origin_column_reshapred)

#see the encoded column
origin_column_encoded


````

````python
#convert the numpy array back to pandas dataframe
origin_dataframe = pd.DataFrame(origin_column_encoded, columns = onehot_encoder.get_feature_names_out(['origin']), index = data.index)

#now it's in pandas
origin_dataframe

````

````python
#Now join it to the original dataframe

data = pd.concat([data, origin_dataframe], axis = 1)

data
````


### **When to use:**
- Label Encoding: For ordinal categories with a natural order
- One-Hot Encoding: For nominal categories without any order relationship
