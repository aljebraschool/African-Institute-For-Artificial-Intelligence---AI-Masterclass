image.png

## **Definition:** Data types specify what kind of data a variable can store and determine how the computer interprets and works with that data. Different types of information need different containers - just like how you'd store liquids in bottles, books on shelves, and clothes in drawers.

**Common Python Data Types:**

*   **Integer (`int`):** Whole numbers (e.g., -5, 0, 100).
    ```python
    count = 15
    ```
*   **Float (`float`):** Decimal numbers (e.g., 3.14, -0.5, 2.0).
    ```python
    temperature = 25.5
    ```
*   **String (`str`):** Sequences of characters (text) enclosed in single or double quotes (e.g., "hello", 'world').
    ```python
    message = "Welcome!"
    ```
*   **Boolean (`bool`):** Represents truth values, either `True` or `False`.
    ```python
    flag = True
    ```
*   **List (`list`):** Ordered collection of items, can contain different data types, mutable (changeable). Enclosed in square brackets `[]`.
    ```python
    numbers = [1, 2, 3, 4, 5]
    ```
*   **Tuple (`tuple`):** Ordered collection of items, can contain different data types, immutable (cannot be changed after creation). Enclosed in parentheses `()`.
    ```python
    coordinates = (10, 20)
    ```
*   **Dictionary (`dict`):** Collection of key-value pairs, where each key is unique. Enclosed in curly braces `{}`.
    ```python
    student = {"name": "Eve", "age": 22, "major": "Computer Science"}
    ```

# 4. Python Math Operations

image.png

## **Definition:** Python math operations are symbols or functions that allow you to perform various mathematical calculations on numbers. Python follows the standard order of operations (PEMDAS - Parentheses, Exponents, Multiplication/Division, Addition/Subtraction).

## Basic Math Operators:

# Addition (+)
```
sum = 5 + 3  # Result: 8

```



# Subtraction (-)


```
difference = 10 - 4  # Result: 6
```



# Multiplication (*)
```
product = 6 * 7  # Result: 42
```



# Division (/)

```
quotient = 20 / 5  # Result: 4.0 (always returns a float)
```

# Integer Division (//)
```
int_quotient = 7 // 2  # Result: 3 (discards the decimal part)
```

# Modulus (%) - returns the remainder after division
```
remainder = 17 % 5  # Result: 2
```

# Exponentiation (**)
```
power = 2 ** 3  # Result: 8 (2 raised to the power of 3)
```

## Compound Assignment Operators:

# Addition assignment (+=)
```
x = 10
x += 5  # Same as x = x + 5, result: 15

```

# Subtraction assignment (-=)

```
y = 8
y -= 3  # Same as y = y - 3, result: 5
```

# Multiplication assignment (*=)
```
z = 4
z *= 2  # Same as z = z * 2, result: 8
```

# Division assignment (/=)
```
a = 9
a /= 3  # Same as a = a / 3, result: 3.0
```

## Built-in Math Functions:

# abs() - absolute value
```
abs(-7)  # Result: 7
```

# round() - rounds to nearest integer or specified decimal places
```
round(3.75)  # Result: 4
round(3.75, 1)  # Result: 3.8 (rounds to 1 decimal place)
```

# min() and max() - find minimum and maximum values
```
min(5, 3, 8, 1)  # Result: 1
max(5, 3, 8, 1)  # Result: 8
```

# 5. Control statement

## **Definition:** Control statements determine the order in which code is executed. They allow your program to make decisions and repeat actions.

image.png

**Common Control Statements:**

*   **If statement:** Executes a block of code only if a certain condition is true.
    ```python
    score = 75
    if score >= 60:
        print("Pass")
    ```
*   **If-else statement:** Executes one block of code if a condition is true, and another block if it's false.
    ```python
    temperature = 15
    if temperature > 20:
        print("It's warm.")
    else:
        print("It's cool.")
    ```
*   **If-elif-else statement:** Allows for multiple conditions to be checked in sequence.
    ```python
    grade = 85
    if grade >= 90:
        print("A")
    elif grade >= 80:
        print("B")
    elif grade >= 70:
        print("C")
    else:
        print("Fail")
    ```
*   **For loop:** Iterates over a sequence (like a list, tuple, or string) or other iterable objects.
    ```python
    fruits = ["apple", "banana", "cherry"]
    for fruit in fruits:
        print(fruit)

    for i in range(5): # Repeats 5 times (from 0 to 4)
        print(i)
    ```
*   **While loop:** Executes a block of code as long as a certain condition is true.
    ```python
    count = 0
    while count < 3:
        print(f"Count is: {count}")
        count += 1
    ```
