# 1. Pandas

image.png

## **Definition:** Pandas is a Python library that makes working with data easy and intuitive. Think of it as an extremely powerful spreadsheet that allows you to manipulate, analyze, and visualize data quickly and efficiently. It is designed to make working with structured (tabular, multidimensional, potentially heterogeneous) and time series data both easy and intuitive.

## Why Pandas?
  - Powerful data manipulation and analysis tools
  - Seamless handling of missing data
  - Intuitive data structures (Series and DataFrame)
  - Easy data importing and exporting from different file formats
  - Integration with other libraries in the data science ecosystem

# 2. Series

image.png

# import pandas library

```
import pandas as pd

```







# Creating a Series from a list



```
fruits = pd.Series(['Apple', 'Banana', 'Orange', 'Mango'])
print(fruits)
```




# Creating a Series with custom index
```
calories = pd.Series([90, 105, 45, 60], index=['Apple', 'Banana', 'Orange', 'Mango'])
print(calories)
```



# 3. DataFrame

image.png

# Creating a DataFrame from a dictionary


```
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 24, 35, 32],
    'City': ['New York', 'Paris', 'Berlin', 'London']
}
df = pd.DataFrame(data)
print(df)
```



# 4. Reading Data

## **Definition:** Pandas provides functions to read data from various file formats into DataFrames.

**Common functions:**
- pd.read_csv(): Read CSV files
- pd.read_excel(): Read Excel files
- pd.read_sql(): Read SQL queries
- pd.read_json(): Read JSON files

# Example:


```
# Reading a CSV file
df = pd.read_csv('data.csv')
```



```
# Reading an Excel file
df = pd.read_excel('data.xlsx', sheet_name='Sheet1')
```




# 5. Data Selection and Indexing


## **Definition:** Pandas offers multiple ways to select and access data within Series and DataFrames.

# Selecting columns

image.png

# Create a simple DataFrame

```
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 24, 35, 32],
    'City': ['New York', 'Paris', 'Berlin', 'London']
}

```



```
df = pd.DataFrame(data)

```
# Select a single column (returns a Series)
```

names = df['Name']
print(names)

```


# Select multiple columns (returns a DataFrame)

```
subset = df[['Name', 'City']]
print(subset)
```




# Selecting rows

image.png

# Selecting rows by position (integer-based)
```
first_row = df.iloc[0]  # First row
specific_rows = df.iloc[[0, 2]]  # First and third rows
slice_rows = df.iloc[1:3]  # Second and third rows

```



image.png

# Selecting rows by label (label-based)



```
df_with_index = df.set_index('Name')
linda_row = df_with_index.loc['Linda']

```



# Combined row and column selection
```
value = df.iloc[0, 1]  # Value at first row, second column
subset = df.iloc[0:2, 1:3]  # First two rows, second and third columns
```



# 6. Data Filtering

# Simple filtering


```
adults = df[df['Age'] >= 30]
new_yorkers = df[df['City'] == 'New York']
```




# Multiple conditions


```
young_europeans = df[(df['Age'] < 30) &
                     ((df['City'] == 'Paris') |
                      (df['City'] == 'Madrid'))]
```



# 7, Basic Operations

# Adding a new column


```
df['Bonus'] = [5000, 7000, 10000, 8000]
```




# Performing calculations on columns

```
df['Total'] = df['Salary'] + df['Bonus']
```



# Applying functions to columns


```
df['Age_in_10_years'] = df['Age'] + 10
```



# Basic statistics
```
average_age = df['Age'].mean()
max_salary = df['Salary'].max()
salary_summary = df['Salary'].describe()  # Comprehensive statistics
```

# Sorting



```
sorted_by_age = df.sort_values('Age')
sorted_by_salary_desc = df.sort_values('Salary', ascending=False)
```
