### Question 1: use arange to create a 2-by-2 array containing the numbers 0–3. Use broadcasting to perform each of the following operations on the original array:


```
a) Cube every element of the array.
b) Add 7 to every element of the array.
c) Multiply every element of the array by 2.

```



````python
#import numpy
import numpy as np

#create a 1D array using 'arange' by default increment is 1
#so you don't need to include that here
num = np.arange(0, 4)

#reshape to 2D array
num = num.reshape(2, 2)

#see your array
print(num)
````

````python
#Cube every element of the array.
num_cube = num ** 3

print(num_cube)
````

````python
#Add 7 to every element of the array.
num_sum = num + 7

print(num_sum)
````

````python
#Multiply every element of the array by 2.
num_by_2 = num * 2

print(num_by_2)
````

### Question 2: Create a 3-by-3 array containing the even integers from 2 through 18. Create a second 3-by-3 array containing the integers from 9 down to 1, then multiply the first array by the second.

````python
#create a even number from 2 to 18.
array1 = np.arange(2, 19, 2)

#since the even numbers in this range is just 9 numbers not 16
#you can reshape using 3 by 3
array1 = array1.reshape(3, 3)

#create an integer array 9 to 1
array2 = np.arange(9, 0, -1)

#reshape it to 3 by 3
array2 = array2.reshape(3, 3)
````

````python
#multiply the two arrays
result = array1 * array2

print(result)
````

### Question 3: Create an array containing the values 1–15, reshape it into a 3-by-5 array, then use indexing and slicing techniques to perform each of the following operations:



```
a) Select row 2.
b) Select column 5.
c) Select rows 0 and 1.
d) Select columns 2–4.
e) Select the element that is in row 1 and column 4.
f) Select all elements from rows 1 and 2 that are in columns 0, 2 and 4.

```



````python
#Create an array containing the values 1–15
num_array = np.arange(1, 16)

#reshape into 3 by 5
num_array = num_array.reshape(3, 5)

print(num_array)
````

````python
# a) Select row 2.
print(num_array[2])
````

````python
# b) Select column 5.
print(num_array[:, 4]) # see all row selected with ':' and use ',' as seperator to move to column
````

````python
# c) Select rows 0 and 1.
print(num_array[0:2]) #select consecutive rows except the last row (2)
````

````python
# d) Select columns 2–4.
print(num_array[:, 2:5])# see all row selected with ':' and use ',' as seperator to move to column before selecting consecutive columns
````

````python
# e) Select the element that is in row 1 and column 4.
print(num_array[1,4])
````

````python
# f) Select all elements from rows 1 and 2 that are in columns 0, 2 and 4.
print(num_array[1:3, [0, 2, 4]]) #select consecutive rows then put the needed columns in list to select them since they're not sequential
````

### Question 4: Perform the following tasks with pandas Series:



```
a) Create a Series from the list [7, 11, 13, 17].
b) Create a Series with five elements that are all 100.0.
c) Create a Series with 20 elements that are all random numbers in the range 0 to 100. Use method describe to produce the Series’ basic descriptive statistics.
```



````python
import pandas as pd

# a) Create a Series from the list [7, 11, 13, 17].
pandas_series = pd.Series([7, 11, 13, 17])

print(pandas_series)
````

````python
# b) Create a Series with five elements that are all 100.0.
pandas_series = pd.Series([100.0] * 5)

print(pandas_series)
````

````python
# c) Create a Series with 20 elements that are all random numbers in the range 0 to 100.
#Use method describe to produce the Series’ basic descriptive statistics.

num = np.random.randint(0, 100, size = 20)

pandas_series = pd.Series(num)

print(pandas_series, end = ' ')
````

### Question 5: Perform the following tasks with pandas DataFrames:



```
a) Create a DataFrame named temperatures from a dictionary of three temperature readings each for 'Maxine', 'James' and 'Amanda'.
b) Recreate the DataFrame temperatures in Part (a) with custom indices using
the index keyword argument and a list containing 'Morning', 'Afternoon'
and 'Evening'.
c) Select from temperatures the column of temperature readings for 'Maxine'.
d) Select from temperatures the row of 'Morning' temperature readings.
e) Select from temperatures the rows for 'Morning' and 'Evening' temperature
readings.

f) Use the describe method to produce temperatures’ descriptive statistics.


```



````python
# a) Create a DataFrame named temperatures from a dictionary of three temperature readings each for 'Maxine', 'James' and 'Amanda'.
named_dict = {'Maxine': 30, 'James': 45, 'Amanda': -10}

temperature = pd.DataFrame(list(named_dict.items()), columns = ['Name', 'Temperature'])

temperature
````

````python
#b) Recreate the DataFrame temperatures in Part (a) with custom indices using
#the index keyword argument and a list containing 'Morning', 'Afternoon' and 'Evening'.
temperature = pd.DataFrame(list(named_dict.items()), columns = ['Name', 'Temperature'], index = ['Morning', 'Afternoon' ,'Evening'])

temperature
````

````python
# c) Select from temperatures the column of temperature readings for 'Maxine'.

temperature.loc['Morning']
````

````python
# d) Select from temperatures the row of 'Morning' temperature readings.
temperature.iloc[0]
````

````python
# e) Select from temperatures the rows for 'Morning' and 'Evening' temperature readings.
temperature.loc[['Morning', 'Evening']]
````

````python
# e) Select from temperatures the rows for 'Morning' and 'Evening' temperature readings.
temperature.iloc[[0, 2]]#thuis gives two rows that not consecutive
````
